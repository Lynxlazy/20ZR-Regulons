```r
suppressPackageStartupMessages({
    library(DESeq2)
    library(biomaRt)
    library(ggplot2)
    library(ggrepel)
    library(lattice)
    library(reshape2)
    library(EnhancedVolcano)
})
options(repr.plot.width = 10, repr.plot.height = 10, repr.plot.res = 100)
```

# CH4_limitation_vs_ctrl.bioreactor

```r
counts <- read.table('CH4_limitation_vs_ctrl.bioreactor_wo_biotype.tsv',header = TRUE, sep = '\t')
head(counts)
metaData <- read.csv('CH4_limitation_vs_ctrl.bioreactor.meta',header = TRUE, sep = '\t')
head(metaData)
corr_mat <- round(cor(counts[,-c(1)]),2)
melted_cor <- melt(corr_mat)
ggplot(data = melted_cor, aes(x=Var1, y=Var2, fill=value)) + 
  geom_tile() +
  geom_text(aes(Var2, Var1, label = value), size = 5) +
  scale_fill_gradient2(low = 'lightblue', high = "orange", midpoint = mean(corr_mat),
                       limit = c(min(corr_mat),1), name="Correlation") +
  theme(axis.title.x = element_blank(),
        axis.title.y = element_blank(),
        panel.background = element_blank())
```

```r
dds <- DESeqDataSetFromMatrix(countData=counts, 
                              colData=metaData, 
                              design=~condition, tidy = TRUE)
dds
dds <- DESeq(dds)
res <- results(dds)
resOrdered <- res[order(res$pvalue),]
write.table(as.data.frame(resOrdered), 
          file="CH4_limitation_vs_ctrl.bioreactor.deg",
          sep = '\t',
          quote = FALSE)
```

```r
EnhancedVolcano(res,
                rownames(res), 
                x = 'log2FoldChange',
                y = 'padj',
                FCcutoff = log2(1.5),
                pCutoff = 0.05,
                title = 'A',
                subtitle = '',
                ylab = bquote(~-Log[10] ~ italic(P)),
                caption = 'FC cutoff = 1.5; P-adj cutoff = 5e-2',
                drawConnectors = T,
                max.overlaps = 20,
                typeConnectors = "open",
                labSize = 5)
```

```r
vsd <- vst(dds, blind=FALSE) 
sampleDists <- dist(t(assay(vsd)))
plotPCA(vsd, intgroup=c("condition"))
```
